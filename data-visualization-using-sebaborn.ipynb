```python
#importing libraries

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
```

```python
data = sns.load_dataset("penguins")
data.head(3)
```

```python
sns.set_theme()
sns.set(rc={'figure.dpi':100})
sns.set(rc = {"figure.figsize":(5,3)})
```

### Scatterplot

```python
sns.scatterplot(x = "bill_length_mm",y="bill_depth_mm",
                data=data,
                hue="island")
```

### Histogram

```python
sns.histplot(x = 'flipper_length_mm',data=data)
```

```python
sns.histplot(y = 'flipper_length_mm',data=data)
```

```python
sns.histplot(data=data, x="flipper_length_mm", binwidth=3)
```

```python
sns.histplot(data=data, x="flipper_length_mm", binwidth=3,kde =True)
```

```python
sns.histplot(data=data, x="flipper_length_mm", hue="species")
```

### Barplot

```python
sns.barplot(x = "species",y="body_mass_g",data=data)
```

```python
sns.barplot(x = "species",y="body_mass_g",data=data,hue="sex")
```

### Boxplot

```python
sns.boxplot(x='island',y="flipper_length_mm",data=data)
```

### Violinplot

```python
sns.violinplot(x='species',y='flipper_length_mm',data= data,hue='sex')
```

### Facetgrid

```python
sns.histplot(data['bill_length_mm'])
```

```python
sns.FacetGrid(data,col="sex",row="island").map(sns.histplot,"bill_length_mm")
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.FacetGrid(data, col="island", row="sex").map(sns.distplot, "flipper_length_mm")
```

### Pairplot

```python
sns.pairplot(data,hue='island',height=6)
```

```python
sns.pairplot(data,hue='sex',height=6,diag_kind ='hist')
```

### Heatmap

```python
sns.heatmap(data.corr(),annot=True)
```
